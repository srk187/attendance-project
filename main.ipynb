```python
import csv
from tkinter import *
from tkinter import ttk
import tkinter as tk
import pandas as pd
from tkinter import messagebox

root = tk.Tk()
root.title('Attendance System')
root.geometry('600x600')
root.resizable(0,0)

#create two frame
frame1 = tk.Frame(root)
frame2 = tk.Frame(root)


#To start frame 1 1st
frame1.pack(fill='both', expand= 1)

#Font Style and size
l = ('consolas',13)

def fun_qr():
    import cv2
    from pyzbar.pyzbar import decode
    import time
    
    cap  = cv2.VideoCapture(0)
    
    cap.set(3, 640)  #3- Width
    cap.set(4, 480)  #4- Height
    used_codes=[]
    
    camera = True
    while camera:
    
        success, frame = cap.read()
    
        for code in decode(frame):
            if code.data.decode('utf-8') not in used_codes:
                print(code.type) 
                print(code.data.decode('utf-8'))
                used_codes.append(code.data.decode('utf-8'))
                time.sleep(5)

                for i in range(len(used_codes)):
                    x,y = map(str, used_codes[i].split(" "))
                    df = pd.DataFrame({'Enrollment_no':[x],'name':[y],'Attendence':['Present']})
                    df.to_csv('attendence.csv',mode='a',index=False,header=False)
                

            elif code.data.decode('utf-8')  in used_codes:
                print('sorry')
                time.sleep(5)
                camera = False

                
            else :
                pass
        
        cv2.imshow('Scan QR Code', frame)
        if cv2.waitKey(1) and 0xFF == ord('q'):
            
            break

    cap.release()
    cv2.destroyAllWindows()

def ch_frame():
    frame2.pack(fill='both', expand = 1)
    frame1.forget()

def back():
    frame1.pack(fill='both', expand= 1)
    frame2.forget()

j=0
r=0
for i in range(100):
    c=str(331221+r)
    Frame(frame1,width=10,height=600,bg='#'+c).place(x=j,y=0)
    j = j + 10
    r = r + 1
Frame(frame1,width=500,height=500,bg='white').place(x=50,y=50)
Button(frame1, width=40, height=2, fg='white', bg='#954535', border=0, text='S T U D E N T   A T T E N D A N C E').place(x=170,y=140)

Button(frame1, width=40, height=2, fg='black', border=0, text='M A N U A L L Y', command=ch_frame).place(x=170,y=280)
Button(frame1, width=20, height=2, fg='black', bg='#CCCCFF', border=0,command=root.destroy, text='E X I T').place(x=240,y=420)
Button(frame1, width=40, height=2, fg='black', border=0, text='Q R   C O D E', command=fun_qr).place(x=170,y=340)
 
#Frame 1 End

#Frame 2 Start

j=0
r=0
for i in range(100):
    c=str(111111+r)
    Frame(frame2,width=10,height=600,bg='#'+c).place(x=j,y=0)
    j = j + 10
    r = r + 1

Frame(frame2,width=500,height=500,bg='white').place(x=50,y=50)

#Variables 
l1_var = StringVar()
l2_var = StringVar()
present = IntVar()

Button(frame2, width=2, height=0, font = l, fg='white', bg='#223344', border=0, command=back, text='<').place(x=50,y=50)

#Enrollment lable
l1 = Label(frame2, text = 'Enrollment No:', bg = 'white')

l1.config(font = l)
l1.place(x=230,y=210)
#Enrollment entry
e1 = Entry(frame2, textvariable=l1_var, width=20, border=0)
e1.config(font=l)
e1.place(x=230,y=240)
#name lable
l2 = Label(frame2, text = 'Name', bg = 'white')
l = ('consolas',13)
l2.config(font = l)
l2.place(x=230,y=280)
#name entry
e2 = Entry(frame2, textvariable=l2_var, width=20, border=0)
e2.config(font=l)
e2.place(x=230,y=310)
Frame(frame2,width=180,height=2,bg='#141414').place(x=230,y=260)
Frame(frame2,width=180,height=2,bg='#141414').place(x=230,y=330)
#Checkbox
Checkbutton(frame2, text="Present", bg='white', font=l, variable=present).grid(padx=230,pady=350)

from PIL import ImageTk, Image
image1 = Image.open("LOG.png")
image2 = ImageTk.PhotoImage(image1)
label1 = Label(frame2,image=image2,border=0,justify=CENTER)
label1.place(x=250,y=50)

#Submit button Functionality
def submit():
    en_no = l1_var.get()
    nam = l2_var.get()
    if(present.get()==1):
        pres = 'Present'
    else:
        pres = 'Absent'

    
    df = pd.DataFrame({'Enrollment_no':[en_no],'name':[nam],'Attendence':[pres]})
    df.to_csv('attendence.csv',mode='a',index=False,header=False)
    e1.delete(0,END)
    e2.delete(0,END)
    
   

#Button
Button(frame2, width=20, height=2, fg='white', bg='#994422', border=0, command=submit, text='S U B M I T').place(x=250,y=420)
Button(frame2, width=20, height=2, fg='white', bg='#994422', border=0, command=root.destroy, text='E X I T').place(x=250,y=480)

#frame 2 end

root.mainloop()


    


```
